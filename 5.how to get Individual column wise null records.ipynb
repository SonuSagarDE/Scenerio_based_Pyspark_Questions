```

df_emp_csv=spark.read.option("nullValue","null").csv("/FileStore/sourcetable/emp.csv",header=True,inferSchema=True)
display(df_emp_csv)
```

```
display(df_emp_csv.filter("EMPNO is null").count())
```

```

from pyspark.sql.functions import count,col,when
# Take null values count at each column
null_cols = df_emp_csv.select([count(when(col(c).isNull(), c)).alias(c) for c in df_emp_csv.columns])
display(null_cols)
# Filter only columns are having null record count..
df_out = null_cols.select([key for (key,value) in null_cols.collect()[0].asDict().items() if value> 0])
df_out.show()
```
