###### First four values are Id,product,vol,city
##### we need to create a data frame with these columns

```
df=spark.read.text("/Filestore/Tables/Sample_Regx.txt")
df.show()
```

```
from pyspark.sql.functions import regexp_replace,explode,split
df1=df.withColumn("New_value",regexp_replace("value","(.*?\\-){4}","$0,")).drop("value")
df2=df1.withColumn("New_value1",explode(split(df1.New_value,'-,')))
df3=df2.withColumn("ID",split(df2.New_value1,'-')[0]) \
        .withColumn("Product",split(df2.New_value1,'-')[1])\
        .withColumn("Vol",split(df2.New_value1,'-')[2])\
        .withColumn("City",split(df2.New_value1,'-')[3])\
        .drop("New_value","New_value1")
                
df3.show()
```
